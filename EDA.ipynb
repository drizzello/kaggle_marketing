```python
place_purchase = ['NumDealsPurchases','NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
# Set the aesthetics for the plots
sns.set(style="whitegrid")

for e in place_purchase:
    plt.figure(figsize=(10, 6))  # Set the figure size for better readability
    sns.countplot(x=e, data=df)  # Specify the source dataframe
    plt.title(f'Count Plot of {e}')  # Set the title for each plot
    plt.show()  # Show the plot
```

```python
off_number = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5','AcceptedCmp2', 'AcceptedCmp1']

sns.set(style="whitegrid")

for e in off_number:
    plt.figure(figsize=(5, 3))  # Set the figure size for better readability
    sns.countplot(x=e, data=df)  # Specify the source dataframe
    plt.title(f'Count Plot of {e}')  # Set the title for each plot
    plt.show()  # Show the plot
           
```

```python
tot_zero = 0
tot_one = 0
for e in off_number:
   zero_count = (df[e] == 0).sum()
   tot_zero += zero_count
   one_count = (df[e] == 1).sum()
   tot_one += one_count

   print(e)
   print(f"0 count:Â {zero_count}")
   print(f"1 count:Â {one_count}")

print(f'Tot not accepted: {tot_zero}')
print(f'Tot accepted: {tot_one}')
```

```python
df['Tot_Amount'] = df['MntWines'] + df['MntSweetProducts'] + df['MntMeatProducts'] + df['MntGoldProds'] + df['MntFruits'] + df['MntFishProducts']
df['Tot_Purchases'] = df['NumWebPurchases'] + df['NumStorePurchases'] + df['NumDealsPurchases'] + df['NumCatalogPurchases'] 
```

```python
# Taking a look at the categorical variables
fig, ax = plt.subplots(1,3, figsize=(15,4))
sns.countplot(data=df, x='Marital_Status', ax=ax[0])
sns.countplot(data=df, x='Education', ax=ax[1])
df_plot = df.groupby(['Marital_Status', 'Education']).size().reset_index().pivot(columns='Marital_Status', index='Education', values=0)
df_plot.apply(lambda x: x/x.sum(), axis=1).plot(kind='bar', stacked=True, ax=ax[2],  colormap='Paired')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))


plt.tight_layout()
fig.show()
```
